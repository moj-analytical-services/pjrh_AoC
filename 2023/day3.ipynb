```python
import re

# part 1

f = open('day3a.txt')
input_file = f.readlines()

line_length = len(input_file[0])-1

dummy_line = ('.' * (line_length))
input_file.insert(0, dummy_line)
input_file.append(dummy_line)

out_val = 0
for i in range(len(input_file)-2):
    i = i + 1

    #print(input_file[i-1])
    #print(input_file[i])
    #print(input_file[i+1])
    
    num_found = re.finditer(r'\b[0-9]+\b', input_file[i])
       
    for x in num_found:
        part_num = int(x.group(0))
        #print(part_num, end="-")
        search_range = slice(max(0, x.span()[0]-1), min(x.span()[1]+1, line_length))
        
        symbol_search = re.compile(r'[^.0123456789]')
        search_str = input_file[i-1][search_range] + input_file[i][search_range] + input_file[i+1][search_range]
        
        if symbol_search.search(search_str):
            out_val = out_val + part_num
            #print("yes")
            continue
            
        #print("no")
            
        
print(out_val)
```

```python
# part 2

import regex

f = open('day3a.txt')
input_file = f.readlines()

line_length = len(input_file[0])-1

dummy_line = ('.' * (line_length))
input_file.insert(0, dummy_line)
input_file.append(dummy_line)

out_val = 0
for i in range(len(input_file)-2):
    i = i + 1
    
    star_found = re.finditer(r'[*]', input_file[i])
       
    for x in star_found:
        search_range = slice(max(0, x.span()[0]-3), min(x.span()[1]+3, line_length))
        #print(search_range)
        
        symbol_search3 = regex.compile(r'\b[0-9]{3}\b') # three digit matches
        symbol_search2 = regex.compile(r'(?<![\n])\b[0-9]{2}\b(?![\n])') # two digit not next to edge
        symbol_search1 = regex.compile(r'(?<![\n][.]{0,1})\b[0-9]\b(?![.]{0,1}[\n])') # one digit not next to edge or with one dot before edge
        search_str = '\n' + input_file[i-1][search_range] + '\n' + input_file[i][search_range] + '\n' + input_file[i+1][search_range] + '\n'
        #print(search_str)
        
        number_list = symbol_search3.findall(search_str) + symbol_search2.findall(search_str) + symbol_search1.findall(search_str)
        
        #print(number_list)
        
        if len(number_list) == 2:
            out_val = out_val + int(number_list[0])*int(number_list[1])
            #print("yes")
            continue
            
        #print("no")
            
        
print(out_val)
```
