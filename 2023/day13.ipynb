```python
# Get data

from aocd import get_data
from collections import Counter
import math
import itertools
import re

# get data
data = get_data(day=13, year=2023)

# data = """#.##..##.
# ..#.##.#.
# ##......#
# ##......#
# ..#.##.#.
# ..##..##.
# #.#.##.#.

# #...##..#
# #....#..#
# ..##..###
# #####.##.
# #####.##.
# ..##..###
# #....#..#"""

data = [string for string in data.split('\n\n')]

#print(*data, sep = "\n---\n")
```

```python
# Part 1

def find_h_mirror(string):
    str_split = string.split('\n')
   
    for rownum in range(1,len(str_split)):
        diff = min(rownum, len(str_split) - rownum)
        if str_split[rownum-diff:rownum] == list(reversed(str_split[rownum:rownum+diff])):
            return rownum
        
    return None
    
    
def find_v_mirror(string):
    transpose = list(map(list, zip(*string.split('\n'))))
    transpose_str = "\n".join((["".join(y) for y in transpose]))
    return find_h_mirror(transpose_str)

h_lines = []
v_lines = []
for item in data:
    h_lines.append(find_h_mirror(item))
    v_lines.append(find_v_mirror(item))

#print(h_lines)
#print(v_lines)

out = sum([x for x in v_lines if x]) + 100*sum([x for x in h_lines if x])

print(out)
```

```python
# Part 2

def find_h_mirror2(string):
    str_split = string.split('\n')
   
    for rownum in range(1,len(str_split)):
        #print(rownum)
        diff = min(rownum, len(str_split) - rownum)
        
        comp1 = ''.join(str_split[rownum-diff:rownum])
        comp2 = ''.join(list(reversed(str_split[rownum:rownum+diff])))
        
        #print(comp1)
        #print(comp2)
        
        flecks = [[x,y] for x,y in zip(comp1, comp2)]
        #print(flecks)
        flips_to_mirror = sum([x!=y for x,y in flecks])
        if flips_to_mirror == 1:
            return rownum
        
    return None
    
    
def find_v_mirror2(string):
    transpose = list(map(list, zip(*string.split('\n'))))
    transpose_str = "\n".join((["".join(y) for y in transpose]))
    return find_h_mirror2(transpose_str)

h_lines = []
v_lines = []
for item in data:
    h_lines.append(find_h_mirror2(item))
    v_lines.append(find_v_mirror2(item))

#print(h_lines)
#print(v_lines)

out = sum([x for x in v_lines if x]) + 100*sum([x for x in h_lines if x])

print(out)
```
