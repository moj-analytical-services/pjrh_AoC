```julia
# Get data

from aocd import get_data
import time

# get data
data = get_data(day=16, year=2023)

# data = r""".|...\....
# |.-.\.....
# .....|-...
# ........|.
# ..........
# .........\
# ..../.\\..
# .-.-/..|..
# .|....-|.\
# ..//.|...."""


data = [string for string in data.split('\n')]

data_width = len(data[0])
data_height = len(data)

#print(*data, sep = "\n")
```

```julia
# Part 1



t0 = time.time()

# the beam_list is a list of beams
# each beam is the history of the tile-paths
# each tile-path element is a list of
# [x coord, y coord, direction in x, direction in y]
beam_list = [[(0,0,1,0)]]

def update_beam(beam, new_tile, beam_list, used_tile_states):
    
    if new_tile in used_tile_states:
        return beam_list, used_tile_states
    elif new_tile[0] >= data_width or new_tile[0] < 0 or new_tile[1] >= data_height or new_tile[1] < 0:
        return beam_list, used_tile_states
    else:
        beam_list = beam_list + [beam + [new_tile]]
        used_tile_states.update({new_tile})
        return beam_list, used_tile_states

def add_next_tile(beam_list, used_tile_states, data):
    
    beam_list_out = []
    for beam in beam_list:
        
        # current state of end of beam
        x, y, dx, dy = beam[-1]
        tile_type = data[y][x]
        #print(tile_type)
        
        if tile_type == ".":
            new_tile = (x+dx, y+dy, dx, dy)
            beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
            
        elif tile_type == "/":
            new_dx, new_dy = [dy*(-1), dx*(-1)]
            new_tile = (x+new_dx, y+new_dy, new_dx, new_dy)
            beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
            
        elif tile_type == "\\":
            new_dx, new_dy = [dy, dx]
            new_tile = (x+new_dx, y+new_dy, new_dx, new_dy)
            beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
    
        elif tile_type == "|":
            if dx == 0: 
                new_tile = (x+dx, y+dy, dx, dy)
                beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
            elif dy == 0:
                for ddy in [1,-1]:
                    new_tile = (x, y+ddy, 0, ddy)
                    beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
                    
        elif tile_type == "-":
            if dy == 0: 
                new_tile = (x+dx, y+dy, dx, dy)
                beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
            elif dx == 0:
                for ddx in [1,-1]:
                    new_tile = (x+ddx, y, ddx, 0)
                    beam_list_out, used_tile_states = update_beam(beam, new_tile, beam_list_out, used_tile_states)
        
    return beam_list_out, used_tile_states


used_tile_states = {beam_list[0][0]}
while len(beam_list) > 0:
    beam_list, used_tile_states = add_next_tile(beam_list, used_tile_states, data)

energized_tiles = set((x, y) for x,y,dx,dy in used_tile_states)
    
print(len(energized_tiles))


t1 = time.time()
print(1000*(t1-t0), "msec")
```
