```python
#part1

from aocd import get_data

# get data
data = get_data(day=8, year=2023)

#data = """LLR
#
#AAA = (BBB, BBB)
#BBB = (AAA, ZZZ)
#ZZZ = (ZZZ, ZZZ)"""

data = data.split("\n")
  
    
lr_string = data[0].replace("\n", "")
print(lr_string)
lr_string_len = len(lr_string)



map_dict = {line[0:3]: {'L': line[7:10], 'R': line[12:15]} for line in data[2:]}
#print(map_dict)

initial_node = "AAA"

def follow_map(map, lrs, initial_node):

    len_lrs = len(lrs)
    index = 0
    node = initial_node
    yield(index+1,node)
    while True:
        node = map[node][lrs[index % len_lrs]]
        yield(index+1, node)
        index += 1

mapper = follow_map(map_dict, lr_string, initial_node)

for i, x in mapper:
    if x == "ZZZ":
        break
    
print(i)
```

```python
#part2
from functools import reduce
import math

initial_p2= [key for key, value in map_dict.items() if key[2]=="A"]
print(initial_p2)

mappers = [follow_map(map_dict, lr_string, initial_n) for initial_n in initial_p2]

s1 = []
for map_gen in mappers:
    for i, x in map_gen:
        if x[2] == "Z":
            break
    s1 = s1 + [i]
    print(i)

s2 = []
for map_gen in mappers:
    for i, x in map_gen:
        if x[2] == "Z":
            break
    s2 = s2 + [i]



print(s1)
print(s2)

print(reduce(lambda x, y: x*y // math.gcd(x, y), s1))
```
