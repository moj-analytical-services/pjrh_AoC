```python
from aocd import get_data
import re
import math
import time


# get data
data = get_data(day=6, year=2023)
data = data.replace("Time:", "").replace("Distance:", "")
data = data.split("\n")

print(data)

def win_check(t, d, hold):
    speed = hold
    time_left = t - hold
    distance_travelled = speed * time_left
    return distance_travelled > d
```

```python
# Part 1

data_p1 = [list(map(int, x.split())) for x in data]

print(data_p1)



wins_all = []
for time, distance in zip(data_p1[0], data_p1[1]):
    mean_speed_to_beat = math.ceil(distance/time)
    wins = 0
    for x in range(mean_speed_to_beat, time):
        win = win_check(time, distance, x)
        if win:
            wins += 1
        else:
            if wins > 0:
                break
    wins_all.append(wins)
    
print("Part 1:")
print(math.prod(wins_all))
```

```python
# Part 2 (same method - slow)
t0 = time.time()

data_p2 = [int(x.replace(" ", ""))for x in data]

print(data_p2)

t_p2 = data_p2[0]
d_p2 = data_p2[1]

mean_speed_to_beat_p2 = math.ceil(d_p2/t_p2)
wins_p2 = 0
for x in range(mean_speed_to_beat_p2, t_p2):
    win = win_check(t_p2, d_p2, x)
    if win:
        wins_p2 += 1
    else:
        if wins_p2 > 0:
            break
    
print("Part 2:")
print(wins_p2)


t1 = time.time()

print(t1-t0)
```

```python
# Part 2 (binary search)
t0 = time.time()

data_p2 = [int(x.replace(" ", ""))for x in data]

print(data_p2)

t_p2 = data_p2[0]
d_p2 = data_p2[1]

mean_speed_to_beat_p2 = math.ceil(d_p2/t_p2)

low = mean_speed_to_beat_p2
high = t_p2
while low <= high:
    mid = low + (high - low)//2
    
    win1 = win_check(t_p2, d_p2, mid-1)
    win2 = win_check(t_p2, d_p2, mid)
        
    
    if (not win1) & (win2):
        lower_win = mid
        break

    if (not win1) & (not win2):
        low = mid + 1
        
    if (win1) & (win2):
        high = mid + 1
        
print(lower_win)    


low = mean_speed_to_beat_p2
high = t_p2
while low <= high:
    mid = low + (high - low)//2
    
    win1 = win_check(t_p2, d_p2, mid-1)
    win2 = win_check(t_p2, d_p2, mid)
        
    
    if (win1) & (not win2):
        higher_win = mid
        break

    if (not win1) & (not win2):
        high = mid - 1
        
    if (win1) & (win2):
        low = mid + 1
        
print(higher_win)    


    
print("Part 2:")
print(higher_win - lower_win)

t1 = time.time()

print(1000*(t1-t0), "msec")
```
