```julia
using Memoize

input = read("day12.txt", String)


row_length = length(split(input, "\n"))
coord_list = [([col_num, row_num], string(char)) for (row_num, row) in enumerate(split(input, "\n")) for (col_num, char) in enumerate(row)]
coord_dict = Dict(coord_list)


function get_neighbours(coord)
    
    check_dirs = [[0,1],[1,0],[0,-1],[-1,0]]
    
    [coord + d for d in check_dirs]
    
end

function get_plot_neighbours(coord, plot_type, coord_dict)
     
    [x for x in get_neighbours(coord) if get(coord_dict, x, "NA") == plot_type]
    
end

visited = Set()
function get_region(start_coord, coord_dict, region = coord_dict[start_coord])
    
    plot_neighbours = get_plot_neighbours(start_coord, region, coord_dict)
    not_visited = setdiff(plot_neighbours, visited)
    global visited = union(visited, plot_neighbours)
    if isempty(not_visited)
        return(union([start_coord], plot_neighbours))
    else
        return(union([start_coord], [get_region(new_coord,coord_dict, region) for new_coord in not_visited]...))
    end

end

```

```julia
allocated_to_region = Set()
regions = []
for (coord, _) in coord_list
    
    if isempty(intersect([coord], allocated_to_region))
        new_region = get_region(coord, coord_dict)
        regions = vcat(regions, [new_region])
        allocated_to_region = union(allocated_to_region, new_region)
    end
    
end

#println(get_region([1,1], coord_dict))

function get_cost(region)
    area = length(region)
    perimeter = 0
    for coord in region
        perimeter = perimeter + 4 - length(intersect(Set(get_neighbours(coord)),Set(region)))
    end
    #println(area, " ", perimeter)
    return(area*perimeter)
end


println("Puzzle 1: ", sum([get_cost(region) for region in regions]))
```

```julia

function get_cost2(region)
    area = length(region)
    #perimeter = Set()
    corners = 0
    for coord in region
        #println(coord)
        #println(region)
        
        
        # Outer corners
        if !((coord + [0,1]) in region) &
        !((coord + [1,0]) in region) &
        !((coord + [1,1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,1]) in region) &
        !((coord + [-1,0]) in region) &
        !((coord + [-1,1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,-1]) in region) &
        !((coord + [-1,0]) in region) &
        !((coord + [-1,-1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,-1]) in region) &
        !((coord + [1,0]) in region) &
        !((coord + [1,-1]) in region)
            corners = corners + 1
        end
        
    
        # Inner corners    
        if ((coord + [0,1]) in region) &
        ((coord + [1,0]) in region) &
        !((coord + [1,1]) in region)
            corners = corners + 1
        end
        
        if ((coord + [0,1]) in region) &
        ((coord + [-1,0]) in region) &
        !((coord + [-1,1]) in region)
            corners = corners + 1
        end
        
        if ((coord + [0,-1]) in region) &
        ((coord + [-1,0]) in region) &
        !((coord + [-1,-1]) in region)
            corners = corners + 1
        end
        
        if ((coord + [0,-1]) in region) &
        ((coord + [1,0]) in region) &
        !((coord + [1,-1]) in region)
            corners = corners + 1
        end
        
        # Inner double corners (will pick up once per side)
        if !((coord + [0,1]) in region) &
        !((coord + [1,0]) in region) &
        ((coord + [1,1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,1]) in region) &
        !((coord + [-1,0]) in region) &
        ((coord + [-1,1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,-1]) in region) &
        !((coord + [-1,0]) in region) &
        ((coord + [-1,-1]) in region)
            corners = corners + 1
        end
        
        if !((coord + [0,-1]) in region) &
        !((coord + [1,0]) in region) &
        ((coord + [1,-1]) in region)
            corners = corners + 1
        end

    end
    #println(perimeter)
    #println(area, " ", corners)
    return(area*corners)
end

println("Puzzle 2: ", sum([get_cost2(region) for region in regions]))
```
