```julia
input = read("day9.txt", String)
#input = "14113"

input_nums = [parse(Int, c) for c in input]

file_info = input_nums[1:2:end]
space_info = input_nums[2:2:end]


files_nested = [repeat([i-1],blocks) for (i,blocks) in enumerate(file_info)]
files_expanded = [i for files in files_nested for i in files]
#println(file_info)
#println(space_info)

i = 1
out = []
while length(files_expanded) > 0
    if mod(i,2) == 1
        index_from = 1
        index_to = min(input_nums[i], length(files_expanded))
        out = vcat(out, splice!(files_expanded,index_from:index_to))
    else
        index_from = max(1, length(files_expanded)-input_nums[i]+1)
        index_to = length(files_expanded)
        out = vcat(out, reverse(splice!(files_expanded,index_from:index_to)))
    end
    i = i + 1
end

#println(out)

checksum = sum([x*(i-1) for (i,x) in enumerate(out)])
println("Puzzle 1: ", checksum)

spaces_out = repeat([[]],length(space_info))
files_moved = []
for (i, file_length) in enumerate(reverse(file_info))
    
    file_id = length(file_info) - i
    
    space_ind = findfirst(x -> x >= file_length, space_info)
    
    if !isnothing(space_ind) && ((space_ind) < (length(file_info) - i + 1))
        spaces_out[space_ind] = vcat(spaces_out[space_ind], repeat([file_id], file_length))
        space_info[space_ind] = space_info[space_ind] - file_length
        files_moved = vcat(files_moved, file_id)
    end
end

out2 = []
file_counter = 0
space_info = vcat(input_nums[2:2:end], 0)
spaces_out = vcat(spaces_out, [[]])
for (file, space) in zip(files_nested, spaces_out)
    
    if length(space) < space_info[file_counter+1]
        space = vcat(space, repeat([0], space_info[file_counter+1] - length(space)))
    end
    
    if file_counter in files_moved
        out2 = vcat(out2, zeros(length(file)), space)
    else
        out2 = vcat(out2, file, space)
    end
    file_counter += 1
end

#println(join([string(Int(n)) for n in out2]))

checksum2 = sum([x*(i-1) for (i,x) in enumerate(out2)])
println("Puzzle 2: ", Int(checksum2))
    
```
